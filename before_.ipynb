### 코드 구현시 체크할 것
1. 명명규칙 확인
    - class : PascalCase
    - method 및 변수 : snake_case
2. 모든 과정은 함수 형태로 선언하고 구현한 후 모듈화
    - 코드 merge 시, 2차 프로젝트 때 쉽게 불러오고 쓸 수 있도록 설정
    - 데이터 확인 및 데이터 정보에 관한 새로운 데이터 생성(컬럼명을 모은 리스트 생성 등) -> DataCheck()
    - 데이터 전처리(결측치 채우기, 컬럼 병합 및 결합, 데이터 인코딩) -> DataPreprocessing()
    - 데이터 시각화(모든 그래프 그리는 기능은 함수로 분리해서 구현) -> DataVisualize()
    - 그 외 코드들도 분리할 수 있는것은 최대한 함수로 분리
